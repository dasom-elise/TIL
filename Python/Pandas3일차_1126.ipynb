```python
data = [[2, np.nan],
       [7,-3],
       [np.nan, np.nan],
       [1,-2]]
df = pd.DataFrame(data,
                 columns=["one","two"],
                 index=["a","b","c","d"])
df
```

```python
# "one"column의 결측값은 one column의 평균으로
# "two"column의 결측값은 two column의 최소값으로 대체해 보아요

one_avg = df["one"].mean(axis=0,skipna=True)
two_avg = df["two"].mean(axis=0,skipna=True)

df["one"] = df["one"].fillna(value = one_avg)
df["two"] = df["two"].fillna(value = two_avg)

df
```

## DataFrame Merge; 데이터프레임 결합하기
### Inner join, Left join, Right join, Outer join

```python
import numpy as np
import pandas as pd

data1 = {
    "학번": [1,2,3,4],
    "이름": ["홍길동","최길동","김연아","강감찬"],
    "학년": [2,3,3,1]
}
data2 = {
    "학번": [1,4,3,5],
    "학과": ["컴퓨터","미술","철학","사회"],
    "학점": [3.0,2.2,4.5,1.7]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
display(df1)
display(df2)
```

```python
df = pd.merge(df1,df2, on = "학번",how = "outer")
df
```

```python
display(pd.merge(df1,df2, on = "학번",how = "outer"))
display(pd.merge(df1,df2, on = "학번",how = "inner"))
display(pd.merge(df1,df2, on = "학번",how = "left"))
display(pd.merge(df1,df2, on = "학번",how = "right"))
```

```python
########### 
## 컬럼명이 다를 경우 조인하려면?
data1 = {
    "학번": [1,2,3,4],
    "이름": ["홍길동","최길동","김연아","강감찬"],
    "학년": [2,3,3,1]
}
data2 = {
    "학생학번": [1,4,3,5],
    "학과": ["컴퓨터","미술","철학","사회"],
    "학점": [3.0,2.2,4.5,1.7]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

display(pd.merge(df1,df2, left_on = "학번", right_on = "학생학번", how = "outer"))
display(pd.merge(df1,df2, left_on = "학번", right_on = "학생학번", how = "inner"))
display(pd.merge(df1,df2, left_on = "학번", right_on = "학생학번", how = "left"))
display(pd.merge(df1,df2, left_on = "학번", right_on = "학생학번", how = "right"))     # 학번+학생학번이 같이나옴
# 출력되는 결과만 명심할것!!
```

```python
## 컬럼이없는경우?
## 인덱스만 있으면?

data1 = {
    "학번": [1,2,3,4],
    "이름": ["홍길동","최길동","김연아","강감찬"],
    "학년": [2,3,3,1]
}
data2 = {
    "학과": ["컴퓨터","미술","철학","사회"],
    "학점": [3.0,2.2,4.5,1.7]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df2.index = [1,2,4,5]     ## column날아가고 index로 바꿈
df1.index = [1,2,3,4]

display(pd.merge(df1,df2, 
                 left_index = True, 
                 right_index = True, 
                 how = "inner"))


# 인덱스와 연결
```
